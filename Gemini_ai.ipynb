```python
import pathlib
import textwrap
import google.generativeai as genai
from IPython.display import display
from IPython.display import Markdown
```

```python
# def to_markdown(text):
#   text = text.replace('*', '  *')
#   return Markdown(textwrap.dedent(text, '> '))
def to_markdown(text):
  text = text.replace('*', '  *')
  # Removing the second argument from textwrap.dedent()
  # and applying it to each line where needed
  text = '\n'.join([line.lstrip('> ') for line in text.splitlines()])
  return Markdown(textwrap.dedent(text))
```
